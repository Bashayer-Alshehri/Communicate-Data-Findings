## Univariate Exploration

> In this section, investigate distributions of individual variables.

#### 1- What is the distribution of trip duration?

```python
#Visualizing the data in a histogram
dur_bins = np.arange(0, df['duration_sec'].max() + 100, 100)
plt.hist(data = df, x = 'duration_sec', bins = dur_bins);

plt.xlim(0, 3000);
plt.style.use('seaborn')
plt.title('Distribution of trip duration', fontsize = 18);
plt.xlabel('Duration of Trip in Seconds');
plt.ylabel('Frequency');
plt.style.use('seaborn');
```

> it's Right-skewed since there's alot of low values at long duration trips.
The peak usage falls in that 250 - 750 second range,about ~4.5 minutes.

> In general, people using the bikes for a short-time durational trips (around 4.5 mins).

#### 2- What is the Start Time of trips by Month?

```python
base_color = sb.color_palette()[0]
sb.countplot(data=df, x='Month', color=base_color)
plt.xlabel('Start Time by Month')
plt.ylabel('Count')
plt.title('Month distribution');
```

> It seems that oct is the most month that people use bikes, the lowest month is jan.

#### 3- What is the distribution of the user type variable??

```python
plt.figure(figsize = [8, 8])
base_color = sb.color_palette()[0]
sb.countplot(data = df, x = 'user_type', color = base_color);
plt.title("Distribution Of The User Type")
plt.xlabel("User Type")
plt.ylabel("Frequency")
plt.show();
```

> It seems that the Subscriber is the most frequent.

```python
df.info()
```

#### 1- The relationship between average bike ride time and months of the year.

```python
# Now check the bike ride duration for each month
plt.figure(figsize = [10,10])
plt.scatter(data = df , x = 'Month' , y = 'duration_sec' , alpha = 0.1 )
plt.title('Monthly Trip Duration')
plt.xlabel('Month')
plt.ylabel('Trip(Secounds)');
```

- There was a constant limit throughout the year and it was about 15,000 seconds.
- June and July had the longest trips during those months.
- November and December had fewer trips that lasted as long.

#### 2- Relationship between seasons and the trip duration in hours?

```python
df['duration_hours'] = df['duration_sec'] / 3600

winter = df.query('Month in ["Dec" , "Jan" , "Feb"]' ).duration_hours.mean()
spring = df.query('Month in ["Mar" , "Apr" , "May"]' ).duration_hours.mean ()
summer = df.query('Month in ["Jun" , "Jul" , "Aug"] ').duration_hours.mean()
fall = df.query('Month in ["Sep" , "Nov" , "Dec"] ').duration_hours.mean()

plt.figure(figsize=[10,8])
plt.bar(['Winter','Spring','Summer','Fall'],[winter , spring , summer , fall] )
plt.title("Relationship Between Seasons And The Trip Duration")
plt.xlabel("Seasons")
plt.ylabel("Duration (hours)");
```

- Almost all of the seasons were close to each other and most of the trips reached 24 mins.

#### 3.What is the relationship between all numeric variables?

```python
plt.figure(figsize = [10, 8])
sb.heatmap(df.corr(), annot = True,cmap = 'Blues')
plt.show()
```

Here in the heatmap plot there are lots of negative correlation and few strong correlation

## Multivariate Exploration

> Create plots of three or more variables .

### 1- The longest trip duration per month for both subscriber and customer?

```python
lo = sb.FacetGrid(data =  df, hue = 'user_type', height = 10)
lo.map(sb.stripplot, 'Month','duration_sec', alpha = 1/5 )
lo.add_legend();
plt.title("The Longest Trip Duration For The User Type")
plt.xlabel("Months")
plt.ylabel("Duration (seconds)");
```

As we can see that customer through all of the months has longer trips duration than subscriber



#### 2- Rush time per month for weekdays?

```python
weekday={0:'Mon', 1:'Tue', 2:'Wed', 3:'Thu', 4:'Fri', 5:'Sat', 6:'Sun'}

df['start_time'] = pd.to_datetime(df['start_time'])
df['weekday'] = df['start_time'].dt.dayofweek.map(weekday)
 
df['weekday'].value_counts()
```

```python
plt.figure(figsize=[10,8])
sb.pointplot(data=df,x='Month',y='duration_hours',hue='weekday', dodge = 0.3 )
plt.title("Rush time per Month For Weekdays")
plt.xlabel("Months")
plt.ylabel("Duration (hours)");
```

Sunday and Saturday in the first months of the year the rush time was approximate about 35 minute to 40 mins.
